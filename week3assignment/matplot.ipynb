```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#line graph
xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,10])
ypoints=np.array([3,10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints,marker='o')
plt.show()#marker make to bold the point ex= "d",'o','*','s'
```

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints,marker='o',linestyle='dotted',color='blue',ms=20,mfc='r')
plt.show()#marker make to bold the point ex= "d",'o','*','s'
```

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints,marker='o')
plt.grid(axis='x')#its shows the table axiz x show the table in x axis if we change to y then the scale is by the y asix
plt.show()#marker make to bold the point ex= "d",'o','*','s'
```

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints,marker='o')
plt.grid()
plt.xlabel('age')#show the x label as age
plt.ylabel('race')#show the y label as race
plt.title('comapring some values')
plt.show()#marker make to bold the point ex= "d",'o','*','s'
```

```python
#bargraph one numerical columns and one is charcater value we use bar graph
```

```python
x=np.array(['A','B','C','D'])
y=np.array([3,4,5,7])
plt.bar(x,y,color='r',width=0.1)#h for horizontal we use in barh it change to horizontal
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([3,4,5,7])
plt.barh(x,y,color='r',height=0.1)#h for horizontal we use in barh it change to horizontal
plt.show()
```

```python
#piechart for fating to check we use piechart
```

```python
x=np.array([12,35,25,25,15])
mylabels=['apple','banana','grapes','cherry','dates']
myexplode=[0.2,0.2,0.2,0,0]
plt.pie(x,labels=mylabels,explode=myexplode)
plt.show()
```

```python
x=np.array([12,35,25,25,15])
mylabels=['apple','banana','grapes','cherry','dates']
mycolors=['black','red','hotpink','purple','yellow']
plt.pie(x,labels=mylabels,colors=mycolors)
plt.legend()#indictes the colorsof labeling
plt.show()
```

```python
#histograms we use it for continuous increase and decrease
```

```python
x=np.array([22,87,5,43,56,74,55,66,77,88,99,51,79,27])
plt.hist(x,bins=[10,30,60,90],ec='red')
plt.title('histogram')
plt.show()
```

```python
#subplot to divide the graph into the given plot line insid the graph
```

```python
#plot1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)#1 row 2 column and this graphshow in 1st place it can also be chnage by putting 1 at 2 and 2 at 1
plt.plot(x,y)

#plot2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)#1 row 2 column at 2 place 
plt.plot(x,y)

plt.show()
```

```python
#plot1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,1,2)#1 row 2 column and this graphshow in 1st place it can also be chnage by putting 1 at 2 and 2 at 1
plt.plot(x,y)

#plot2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,1,1)#1 row 2 column at 2 place #1 columns and 2 row 
plt.plot(x,y)

plt.show()
```
