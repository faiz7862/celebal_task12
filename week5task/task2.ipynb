```python
# Assuming 'HeartDisease' column indicates presence (1) or absence (0) of heart disease
heart_disease_rows = df[df['heart_disease'] == 1]
print(heart_disease_rows)
```

```python
# Assuming 'Hypertension' column indicates presence (1) or absence (0) of hypertension
hypertension_no_heart_disease_rows = df[(df['hypertension'] == 1) & (df['heart_disease'] == 0)]
print(hypertension_no_heart_disease_rows)
```

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Option 1: Drop rows with missing values
cleaned_df_drop = df.dropna()
print("Data after dropping rows with missing values:\n", cleaned_df_drop)

# Option 2: Fill missing values with mean or other appropriate values
# For numerical columns, you can fill missing values with the column mean
for column in df.select_dtypes(include=np.number).columns:
    df[column].fillna(df[column].mean(), inplace=True)

# For categorical columns, you can fill missing values with the mode (most frequent value)
for column in df.select_dtypes(include='object').columns:
    df[column].fillna(df[column].mode()[0], inplace=True)

print("Data after filling missing values:\n", df)
```

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('healthcare.csv')

# Display the first few rows of the dataset to understand its structure
print("First few rows of the dataset:\n", df.head())

# Display rows that have heart disease
heart_disease_rows = df[df['heart_disease'] == 1]
print("Rows with heart disease:\n", heart_disease_rows)

# Display rows that have hypertension and no heart disease
hypertension_no_heart_disease_rows = df[(df['hypertension'] == 1) & (df['heart_disease'] == 0)]
print("Rows with hypertension and no heart disease:\n", hypertension_no_heart_disease_rows)

# Check for missing values in the dataset
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Option 1: Drop rows with missing values
cleaned_df_drop = df.dropna()
print("Data after dropping rows with missing values:\n", cleaned_df_drop)

# Option 2: Fill missing values with mean or other appropriate values
# For numerical columns, you can fill missing values with the column mean
for column in df.select_dtypes(include=np.number).columns:
    df[column].fillna(df[column].mean(), inplace=True)

# For categorical columns, you can fill missing values with the mode (most frequent value)
for column in df.select_dtypes(include='object').columns:
    df[column].fillna(df[column].mode()[0], inplace=True)

print("Data after filling missing values:\n", df)
```
